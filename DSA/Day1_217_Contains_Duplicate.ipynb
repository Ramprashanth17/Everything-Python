# 217. Contains Duplicate

Given an integer array nums, return true if any value appears at least twice in the array, and return false if every element is distinct.

### Approach-1: Using Brute Force or Naive Method

The question says that given an array of numbers, if a value is repeated(duplicate!), we've to return True.

Basically, we can use a nested for loop over here to solve this. Outer loop keeps hold of the element we need to compare with other elements and in the inner loop, we'll try and see if an element matches with the element we've on hold, if there's a match, then it's a duplicate and return True, otherwise,keep on going with the other elements.

We can sort the values first, so that in case of duplicate values being side by side, we can return faster.

**Drawbacks:** We are employing a nested loop which is of O(n^2) complexity, not ideal as the input size grows the computing time also grows.


```python
### Approach-1: Brute Force Method ####

class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        n = len(nums)
        # Outer loop picks the first number to compare
        for i in range(n):
            # Inner loop scans all subsequent numbers
            for j in range(i + 1, n):
                # Comparison
                if nums[i] == nums[j]:
                    return True
        return False
```

### Approach-2: Using a set

Set is a data structure in Python which is unordered and doesn't allow duplicates. So, all we need to do is to compare the length of the array before conversion to set and length of the array after conversion. If they're unequal then there are duplicates.

**Pros:** Faster to implement than the Naive Method, Time complexity is O(n) and the space complexity is also O(n) ***[converting a list to a set]***

**Cons**: What if the first 2 elements are duplicates? Our method still does the set conversion and gives out a value, in worst case scenario our apporoach makes sense but in best case scenario? O(n) isn't good, same with space complexity, as we can further optimize it!


```python

class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        return len(set(nums)) != len(nums)
```

### Approach-3: The optimized approach

In this optimized approach we'll address the best and average case scenario ***"What if, if the first two elements are duplicates?"***

We'll be using set() again, but with a slight twist! We'll initialize an empty set first and we'll employ a for loop to iterate through elements in the array, if not present in the set, then add the element to the set. If there are duplicates in the array, then it can't be added to the set, as that element is already present in it and as you know, set won't allow duplicates, hence return True. That's it!!

**Pros:**: Optimized version of the solution, in the worst case time and space complexity is O(n), but in best and average cases, the space complexity is O(1) and time complexity is O(1) as it requires only one operation.



```python

class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        answer = set()
        for num in nums:
            if num not in answer:
                answer.add(num)
            else:
                return True
        return False
```
