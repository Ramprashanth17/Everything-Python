# Reading data from different file formats

**CSVs**

```python
import csv

source_file = open("/content/sample_data/california_housing_test.csv", "r")
csv_reader = csv.DictReader(source_file)

print(csv_reader.fieldnames) # returns the column names
print()
for i in range(5):
  print(next(csv_reader))
```

**Reading TSVs or TXTs**

```python
import csv

tsv_source_file = open("/content/Austin_Column_Corrected.tsv", "r")
tsv_reader = csv.DictReader(tsv_source_file, delimiter = '\t') #Remains same for both tab and txt files

#Field Names in the file
print(tsv_reader.fieldnames)
print()
for i in range(5):
  print(next(tsv_reader))
```

**Reading from xlsx**

```python
!pip install openpyxl

from openpyxl import load_workbook

import csv

src_wb = load_workbook(filename = "/content/mapping (1).xlsx")
print(src_wb.sheetnames) # returns sheetnames in the excel file

print()

for sheet_num, sheet_name in enumerate(src_wb.sheetnames):
  current_sheet = src_wb[sheet_name]
  print(sheet_name)

  output_file = open("xlsx_" + sheet_name + ".csv", "w")
  output_writer = csv.writer(output_file)

  # loop through every row in the sheet

  for row in current_sheet.iter_rows():
    row_cells = [] # to store value of cells in each row
    for cell in row:
      print(cell, cell.value)
      row_cells.append(cell.value)

      output_writer.writerow(row_cells)

  output_file.close()


```

**Fixed Width**

```python
import csv

filename = "ghcnd-stations"

src_file = open(filename+".txt", "r")

stations_list = src_file.readlines()

output_file = open(filename+".csv","w")

output_writer = csv.writer(output_file)

headers = ["ID","LATITUDE","LONGITUDE","ELEVATION","STATE","NAME","GSN_FLAG",
           "HCNCRN_FLAG","WMO_ID"]

output_writer.writerow(headers)

for line in stations_list:
  new_row = []
  # ID: positions 1-11
  new_row.append(line[0:11])

    # LATITUDE: positions 13-20
  new_row.append(line[12:20])

    # LONGITUDE: positions 22-30
  new_row.append(line[21:30])

    # ELEVATION: positions 32-37
  new_row.append(line[31:37])

# STATE: positions 39-40
  new_row.append(line[38:40])

    # NAME: positions 42-71
  new_row.append(line[41:71])

    # GSN_FLAG: positions 73-75
  new_row.append(line[72:75])

    # HCNCRN_FLAG: positions 77-79
  new_row.append(line[76:79])

    # WMO_ID: positions 81-85
  new_row.append(line[80:85])

  output_writer.writerow(new_row)

output_file.close()


# Reading the generated csv
csv_file = open("/content/ghcnd-stations.csv","r")
csv_reader = csv.reader(csv_file)
for i in range(5):
  print(next(csv_reader))
```
